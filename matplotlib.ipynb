# BASICS OF MATPLOTLIB

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x, y)
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x, y, label="sine wave")
plt.title("yo ho sine wave")
plt.xlabel("ye he x axis")
plt.ylabel("ye he y axis")
plt.legend()
plt.show()
```

```python
#bar chart
#bar chart
import matplotlib.pyplot as plt
categories=['A', 'B', 'C', 'D']
values=[10, 20, 30, 40]
plt.bar(categories, values, color=list('rgbk'))
plt.title("Bar Chart")
plt.xlabel("yo ho categories")
plt.ylabel("yo ho Values")
plt.show()


```

```python
import seaborn as sns
```

```python
#histogram and kde plot
data= np.random.randn(1000)
sns.histplot(data, kde=True, color='blue')
plt.title("this is histogram")
plt.show()
```

```python
#heat-map
matrix=np.random.rand(5, 5)
matrix
```

```python
sns.heatmap(matrix, cmap='coolwarm', annot=True )
plt.title("yo ho heatmap")
plt.show()
```

```python
#pairplot
sns.set_style("whitegrid")
df=sns.load_dataset("iris")
sns.pairplot(df, hue="species", palette="Set1", markers=["o", 's', 'd'])
plt.show()
```
